**Normalizando variables de autos usados**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("04-auto.csv")

df["longitud_normalizada"] = np.round(df["longitud"]/df["longitud"].max(), 4)
df["ancho_normalizado"] = np.round(df["ancho"]/df["ancho"].max(), 4)
df["altura_normalizado"] = np.round(df["altura"]/df["altura"].max(), 4)

display(df[["id", "longitud_normalizada", "ancho_normalizado"]].head())
```

**Binning Uniforme**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("04-auto.csv")

bins = np.linspace(df["potencia_hp"].min(), df["potencia_hp"].max(), num=4)
etiquetas = ["Baja", "Media", "Alta"]
df["potencia_hp_agrupada"] = pd.cut(
    df["potencia_hp"], bins=bins, labels=etiquetas, include_lowest=True
)

display(df[["id", "potencia_hp", "potencia_hp_agrupada"]].head())
```

**Binning mega manual**

```python
import pandas as pd

df = pd.read_csv("04-auto.csv")

i = 0
while(i < len(df)) :
  if(df.loc[i, "potencia_hp"] <= 100) :
    df.loc[i, "potencia_hp_agrupada"] = "Baja"
  elif(df.loc[i, "potencia_hp"] <= 200) :
    df.loc[i, "potencia_hp_agrupada"] = "Media"
  else :
    df.loc[i, "potencia_hp_agrupada"] = "Alta"
  i += 1

display(df[["id", "potencia_hp", "potencia_hp_agrupada"]].head())
```

**Consolidado**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("04-auto.csv")

# Normalizacion de longitud, ancho y altura
df["longitud_normalizada"] = np.round(df["longitud"]/df["longitud"].max(), 4)
df["ancho_normalizado"] = np.round(df["ancho"]/df["ancho"].max(), 4)
df["altura_normalizado"] = np.round(df["altura"]/df["altura"].max(), 4)

# Binning uniforme de potencia_hp
bins = np.linspace(df["potencia_hp"].min(), df["potencia_hp"].max(), num=4)
etiquetas = ["Baja", "Media", "Alta"]
df["potencia_hp_agrupada"] = pd.cut(
    df["potencia_hp"], bins=bins, labels=etiquetas, include_lowest=True
)

# One-Hot Encoding para tipo_combustible y aspiracion
df_dummy_tc = pd.get_dummies(
    df["tipo_combustible"], prefix="tipo_combustible", dtype=int
)
df = pd.concat([df, df_dummy_tc], axis=1)
df_dummy_as = pd.get_dummies(df["aspiracion"], prefix="aspiracion", dtype=int)
df = pd.concat([df, df_dummy_as], axis=1)

df.to_csv("04-auto-v2-limpio-preparado.csv", index=False)
```

**Ejercicio 3**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("04-laptop.csv")

# Normalizacion Min-Max frecuencia_cpu
df["frecuencia_cpu_normalizada"] = np.round((
    df["frecuencia_cpu"] - df["frecuencia_cpu"].min()
) / (
    df["frecuencia_cpu"].max() - df["frecuencia_cpu"].min()
), 4)

# Binning uniforme precio
bins = np.linspace(df["precio"].min(), df["precio"].max(), num=4)
etiquetas = ["Bajo", "Medio", "Alto"]
df["precio_agrupado"] = pd.cut(
    df["precio"], bins=bins, labels=etiquetas, include_lowest=True
)

# One-Hot Encoding para pantalla
df_dummy_p = pd.get_dummies(df["pantalla"], prefix="pantalla", dtype=int)
columnas = {"pantalla_IPS Panel" : "pantalla_ips_panel"}
df_dummy_p = df_dummy_p.rename(columns=columnas)
df = pd.concat([df, df_dummy_p], axis=1)

df.to_csv("04-laptop-v2-limpio-preparado.csv")
```
