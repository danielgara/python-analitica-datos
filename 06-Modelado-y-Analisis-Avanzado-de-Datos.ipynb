**Ejemplo de codigo para regresion**

```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor

datos = {
  "pais": ["Turquía", "Hungría", "Estados Unidos",
           "Nueva Zelanda", "Australia", "Dinamarca"],
  "GDP": [28384, 31008, 60236, 42404, 48698, 55938],
  "satisfaccion": [5.5, 5.6, 6.9, 7.3, 7.3, 7.6]
}

df = pd.DataFrame(datos)
display(df)

X = df["GDP"].to_frame().values
Y = df["satisfaccion"].values

modelo = KNeighborsRegressor(n_neighbors=1)
modelo.fit(X, Y)

X_gdp_francia = [[42026]]
satisfaccion_predicha = modelo.predict(X_gdp_francia)

# la original era 6.5
print("Satisfacción: "+str(satisfaccion_predicha[0]))
```

**Ejemplo de codigo para clasificacion**

```python
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

df = pd.DataFrame({
  "edad" : [16, 17, 18 , 25, 30, 35],
  "cortadas" : [5, 6, 7, 0 , 1 , 0],
  "emo" : [1 , 1 , 1 , 0 , 0 , 0]
})

X = df[["edad", "cortadas"]].values
Y = df["emo"].values

modelo = KNeighborsClassifier(n_neighbors=3)
modelo.fit(X, Y)

X_nueva_persona = [[19, 4]]

prediccion = modelo.predict(X_nueva_persona)
if prediccion[0] == 1 :
  print("La persona es emo.")
else :
  print("La persona no es emo.")
```

**Ejercicio 4**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("06-auto.csv")

tam_motor_max = df["tamano_motor"].max()
potencia_hp_max = df["potencia_hp"].max()
df["tamano_motor"] = df["tamano_motor"] / tam_motor_max
df["potencia_hp"] = df["potencia_hp"] / potencia_hp_max

X_ent = df[["tamano_motor", "potencia_hp"]].values
Y_ent = df["precio"].values

modelo = LinearRegression()
modelo.fit(X_ent, Y_ent)

X_pred = [[202/tam_motor_max, 110/potencia_hp_max]]
Y_pred = modelo.predict(X_pred)

print("Intercepto: "+str(modelo.intercept_))
print("Coeficiente: "+str(modelo.coef_))
print("Prediccion: "+str(Y_pred[0]))
```

**In-sample evaluation**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

df = pd.read_csv("06-auto.csv")

X_ent = df["tamano_motor"].to_frame().values
Y_ent = df["precio"].values

modelo = LinearRegression()
modelo.fit(X_ent, Y_ent)

X_pred = X_ent
Y_pred = modelo.predict(X_pred)

mse = mean_squared_error(Y_ent, Y_pred)
r2 = r2_score(Y_ent, Y_pred)
rmse = np.sqrt(mse)

print("In-sample MSE: "+str(round(mse, 2)))
print("In-sample R^2: "+str(round(r2, 4)))
print("Error promedio: "+str(round(rmse, 2)))
```

**Out of sample Evaluation**

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import numpy as np

df = pd.read_csv("06-auto.csv")

X = df["tamano_motor"].to_frame().values
Y = df["precio"].values

# Divide en entrenamiento y prueba (80% ent, 20% prueba)
X_ent, X_test, Y_ent, Y_test = train_test_split(X, Y, test_size=0.2)

modelo = LinearRegression()
modelo.fit(X_ent, Y_ent)

X_pred = X_test
Y_pred = modelo.predict(X_pred)

mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)
rmse = np.sqrt(mse)

print("In-sample MSE: "+str(round(mse, 2)))
print("In-sample R^2: "+str(round(r2, 4)))
print("Error promedio: "+str(round(rmse, 2)))
```

**Grid Search**

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, r2_score

df = pd.read_csv("06-auto.csv")

X_ent = df["tamano_motor"].to_frame().values
Y_ent = df["precio"].values

# Definir el modelo base
modelo_base = RandomForestRegressor(random_state=42)

# Definir el grid de hiperparametros a probar
param_grid = {
    "n_estimators" : [50, 100, 200],
    "max_depth" : [3, 5, None],
    "min_samples_split" : [2, 5, 10]
}

# Configurar Grid Search con validacion cruzada (5 folds)
grid_search = GridSearchCV(
    estimator=modelo_base,
    param_grid=param_grid,
    scoring=make_scorer(r2_score),
    cv=5,
    verbose=2,
)

# Ajustar Grid Search
grid_search.fit(X_ent, Y_ent)

# Mostrar mejores parametros
print("Mejores hiperparametros encontrados: ")
print(grid_search.best_params_)

# Usar el mejor modelo para predecir un nuevo valor
mejor_modelo = grid_search.best_estimator_
X_pred = [[202]]
Y_pred = mejor_modelo.predict(X_pred)
print("El precio predicho para motor 202 es : "+str(Y_pred[0]))
```
