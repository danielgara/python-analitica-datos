**Retorno de informacion : dtypes , value_counts() , unique().tolist() , mean() , filtro**

```python
import pandas as pd

df = pd.read_csv("02-auto.csv", encoding="latin1")
s_columna_altura = df["altura"]

print(s_columna_altura.dtypes)
print("")
print(s_columna_altura.value_counts())
print("")
print(s_columna_altura.unique().tolist())
print("")
print(s_columna_altura.mean())
print("")
print(s_columna_altura[s_columna_altura > 59])
```

**Ejercicio 5**

```python
import pandas as pd

df = pd.read_csv("02-laptop.csv", encoding="latin1")
s_columna_precio = df["Price"]

print(s_columna_precio.dtypes)
print("")
print(s_columna_precio.min())
print("")
print(s_columna_precio.max())
print("")
print(s_columna_precio.iloc[0])
print("")
print(s_columna_precio.value_counts())
print("")
print(s_columna_precio[s_columna_precio < 800])
print("")
print(s_columna_precio[s_columna_precio > 2000].sort_values())
```
