**Ejercicio 1**

```python
import pandas as pd

edades = [10, 15, 15, 20, 25]
df = pd.DataFrame(edades, columns=['edad'])
s_edades = df["edad"]

# Medidas de tendencia central
promedio = s_edades.mean()
mediana = s_edades.median()
moda = s_edades.mode()[0]

# Medidas de dispersión
rango = s_edades.max() - s_edades.min()
varianza = s_edades.var()
desviacion_estandar = s_edades.std()

print("Promedio: "+str(promedio))
print("Mediana: "+str(mediana))
print("Moda: "+str(moda))
print("Rango: "+str(rango))
print("Varianza: "+str(varianza))
print("Desviación estándar: "+str(desviacion_estandar))
```

**Boxplot - Grafica**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

alturas = [140, 150, 155, 160, 165, 170, 175, 220]

Q1 = np.percentile(alturas, 25)
Q2 = np.median(alturas)
Q3 = np.percentile(alturas, 75)
IQR = Q3 - Q1
limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

sns.boxplot(y=alturas)
plt.axhline(limite_superior, color='purple', linestyle='--')
plt.axhline(limite_inferior, color='purple', linestyle='--')
plt.ylabel("Altura")
plt.title("Boxplot de Alturas")
plt.grid(axis='x')
plt.show()
```

**DataFrameGroupBy V2 - Visualizacion**

```python
import pandas as pd

df = pd.DataFrame({
    'pais' : ['Peru', 'Peru', 'Colombia', 'Colombia', 'Colombia'],
    'ciudad' : ['Lima', 'Lima', 'Bogota', 'Bogota', 'Medellin'],
    'ventas' : [100, 150, 200, 180, 120],
})

display(df.head())

dfgb = df.groupby(['pais', 'ciudad'])
df_grupo_p_l = dfgb.get_group(('Peru', 'Lima'))
display(df_grupo_p_l)

df_grupo_c_b = dfgb.get_group(('Colombia', 'Bogota'))
display(df_grupo_c_b)

df_grupo_c_m = dfgb.get_group(('Colombia', 'Medellin'))
display(df_grupo_c_m)

df_ventas_suma = dfgb.sum()
display(df_ventas_suma)
```

**Pivot - Ejemplo con un index - Codigo**

```python
import pandas as pd

datos = {
    "marca" : ["Toyota", "Toyota", "Ford", "Ford", "Mazda", "Mazda"],
    "num_puertas" : [2, 4, 2, 4, 2 , 4],
    "precio" : [18000, 19000, 17000, 16000, 16500, 15000]
}

df = pd.DataFrame(datos)
display(df)

df_pivot = df.pivot(index="marca", columns="num_puertas", values="precio")
display(df_pivot)
```

**pivot_table**

```python
import pandas as pd

df = pd.read_csv("05-auto.csv")
df_filtrado = df[["marca", "numero_puertas", "precio"]]
df_pivot = df_filtrado.pivot_table(
    index="marca",
    columns="numero_puertas",
    values="precio",
    aggfunc="mean")

display(df_pivot)
```

**Heatmao de correlaciones**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("05-auto.csv")
df_numerico = df.select_dtypes(include=["number"])
df_corr = df_numerico.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(df_corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.tight_layout()
plt.show()
```
