**Reemplazo por promedio**

**Reemplazo por el valor mas frecuente**

**Problema**

```python
import pandas as pd
import numpy as np

df = pd.read_csv("03-auto-v2-columnas.csv")
# Reemplazoo de datos faltantes por NaN
df = df.replace("?", np.nan)

# Reemplazo de marca faltantes por valor estandar
df["marca"] = df["marca"].replace(np.nan, "alfa-romero")

# Reemplazo de diametro_cilindro faltantes por promedio
promedio_dc = np.round(df["diametro_cilindro"].astype("float").mean(), 2)
df["diametro_cilindro"] = df["diametro_cilindro"].replace(np.nan, promedio_dc)

# Reemplazo de numero_puertas faltantes por valor mas frecuente
mas_frecuente_np = df["numero_puertas"].value_counts().idxmax()
df["numero_puertas"] = df["numero_puertas"].replace(np.nan, mas_frecuente_np)

# Borrado de filas sin precio
df = df.dropna(subset=["precio"])

# Guardado de la nueva version
df.to_csv("03-auto-v3-sin-faltantes.csv", index=False)
```

**Consolidado**

```python
import pandas as pd

df = pd.read_csv("03-auto-v3-sin-faltantes.csv")

# Estandarizar carroceria
df["estilo_carroceria"] = df["estilo_carroceria"].replace("CONVERTIBLE", "convertible")

# Estandarizar unidades a kilometros
df["consumo_carretera_mpg"] = df["consumo_carretera_mpg"]*1.60934
df = df.rename(columns={"consumo_carretera_mpg": "consumo_carretera_kpg"})
df["consumo_ciudad_mpg"] = df["consumo_ciudad_mpg"]*1.60934
df = df.rename(columns={"consumo_ciudad_mpg": "consumo_ciudad_kpg"})

# Modificar tipo de columna a int
df["consumo_carretera_kpg"] = df["consumo_carretera_kpg"].astype("int")
df["consumo_ciudad_kpg"] = df["consumo_ciudad_kpg"].astype("int")

df.to_csv("03-auto-v4-formateado.csv", index=False)
```
