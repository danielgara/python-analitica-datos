**Condicionales**

```python
nota = float(input("Ingresa tu nota: "))

if(nota >= 3.0):
  print("Ganaste")

print("Fin programa")
```

**Condicionales - Otras estructuras**

```python
temperatura = float(input("Ingresa la temperatura: "))

if(temperatura >= 26):
  print("Calor")
elif(temperatura >= 15):
  print("Templado")
else:
  print("Frio")

print("Fin programa")
```

**Ejercicio 2**

```python
nombre1 = input("Ingrese el nombre del primer candidato: ")
votos1 = int(input("Ingrese la cantidad de votos del primer candidato: "))
nombre2 = input("Ingrese el nombre del segundo candidato: ")
votos2 = int(input("Ingrese la cantidad de votos del segundo candidato: "))

if(votos1 > votos2):
  print("El ganador es: "+nombre1)
elif(votos2 > votos1):
  print("El ganador es: "+nombre2)
else:
  print("Empate")
```

**Ciclo While**

```python
i = 1

while(i < 5):
  print(i)
  i = i+1

print("Fin programa")
```

**Ciclo For**

```python
for i in range(1,6,1):
  print(i)
```

**Ciclo infinito con parada**

```python
while(True):
  centro = input("Ingrese el nombre del mejor centro de EAFIT: ")
  if(centro == "Urbam"):
    print("Te veo bien")
    break
  else:
    print("Te veo mal")

print("Fin programa")
```

**Listas y Diccionarios: Acceder, modificar , agregar y eliminar**

```python
equipos = ["Ascensor", "Puente grua", "Invernadero"]

print(equipos[0])
equipos[1] = "Banda"
equipos.append("Minisecuenciador")
equipos.pop(0)
print(equipos)
```

```python
ciudad = {"nombre": "Medellin", "fundacion": 1616}

print(ciudad["fundacion"])
ciudad["nombre"] = "KarolG"
ciudad.pop("fundacion")
ciudad["elevacion"] = 1495
print(ciudad)
```

**Ejercicio 4**

```python
cohete1 = {"nombre" : "Falcon Heavy", "capacidad":63.8}
cohete2 = {"nombre" : "New Glenn", "capacidad":45}
cohete3 = {"nombre" : "Starship", "capacidad":150}

cohetes = []
cohetes.append(cohete1)
cohetes.append(cohete2)
cohetes.append(cohete3)

for cohete in cohetes:
  print(cohete["nombre"] + " - " + str(cohete["capacidad"]))
```

**Funciones: Flujo de ejecucion**

```python
def doble_edad(edad) :
  return edad*2

edad = 20
nueva_edad = doble_edad(edad)
print(nueva_edad)

edad2 = 30
nueva_edad2 = doble_edad(edad2)
print(nueva_edad2)
```
